## Distribution of Dataset

```python
sns.boxplot(x='Gender', y='Sleep Duration', data= df, width= 0.3, linewidth = 1)
plt.title('Visualizing The Sleep Duration Of People Using Box Plots')
plt.show()
```

For males, the mean sleep duration is approximately 7.5 hours, with the majority falling within the range of 6 to 8 hours. The box plot indicates a tight distribution, suggesting consistency in sleep patterns. The upper whisker extends to 8 hours, indicating the maximum sleep duration observed, while the lower whisker reaches 6 hours, representing the minimum sleep duration.

In contrast, females exhibit a slightly lower mean sleep duration of around 7 hours. The box plot for females shows a wider distribution, ranging from 6 to 8.5 hours. The upper whisker extends to 8.5 hours, indicating the maximum observed sleep duration, while the lower whisker reaches 6 hours.

However, the analysis suggests that the variability in sleep duration is more pronounced among females, with some individuals experiencing both shorter and longer sleep durations.

```python
sns.boxplot(df, y='Age', x= 'Physical Activity Level', width=0.35)
plt.title('Showcasing the physical activity levels of different age groups using Box and whisker plots')
plt.show()
```

The mean age for most activity levels is concentrated within the range of 30 to 45 years. This suggests that individuals within this age group tend to have similar average physical activity levels.

The longest whisker is observed for activity levels 30 and 75. This indicates that these specific activity levels have a higher variability in ages. The spread of ages for these activity levels is wider, potentially suggesting a more diverse age distribution and varied physical activity behaviors.

For activity level 30 and 75, where the mean age is around 40, the presence of the longest whisker may indicate that individuals within these activity levels exhibit a broader age range.

```python
sns.lineplot(df,x='Age', y='Stress Level', color='Red')
plt.title('Scatterplot Displaying Trends Of Stress Level W.R.T AGE')
plt.show()
```

The line plot reveals a noteworthy pattern in stress levels concerning age. Between the ages of 35 and 45, there is a decreasing trend, with stress levels declining from 8 to 6. Following this, between ages 45 and 55, stress levels remain consistently stable. This suggests that individuals experience a decrease in stress during their mid-thirties to mid-forties, followed by a period of sustained stress levels in the subsequent decade.

# Correlation Matrix


```python
correlation_coefficients = df.corr()
correlation_data =df.corr()
correlation_data
```

```python
sns.heatmap(correlation_data, annot=True, cmap="magma")
plt.title("Data Insights: Variable Associations in Heatmap")
plt.show()
```

The high correlation coefficients are 0.88, 0.77 and 0.67 that suggests strong positive relationships.
These relations are between Sleep Duration - Quality of Sleep, Daily Steps - Physical Activity Level and Stress Level - Heart Rate

A strong positive correlation of 0.88 between sleep duration and the quality of sleep indicates a robust relationship. This suggests that as sleep duration increases, the quality of sleep tends to improve, and vice versa. These variables are positively associated, meaning that individuals who sleep longer are more likely to report higher sleep quality.

The positive correlation indicates that as the number of daily steps increases, there tends to be a corresponding improvement in the physical quality of sleep, and vice versa.This finding may have significant implications for health and well-being, as it suggests a potential connection between physical activity and sleep quality. Individuals who engage in higher levels of daily steps may experience better physical sleep quality, which aligns with existing research highlighting the positive impact of physical activity on sleep patterns.

The positive correlation of 0.67 between stress level and heart rate implies a moderate association between these variables. This suggests that higher stress levels are moderately associated with higher heart rates. It's important to note that correlation does not imply causation, but in this case, the correlation suggests a connection between increased stress and elevated heart rates.

```python
sns.displot(df, x='Daily Steps', hue='BMI Category', kind='kde')
plt.title('Visualizing the Density of BMI Category Levels in Relation to Daily Step Counts with a Density Plot')
plt.show()
```

A concentration of daily steps between 6000 to 9000 being associated with the "Normal" BMI category suggests that individuals within this steps range are more likely to have a BMI within the normal weight range. This aligns with general health recommendations emphasizing the positive impact of moderate physical activity on maintaining a healthy weight.

The density plot indicates that daily steps falling between 4000 to 6000 are associated with the "Overweight" BMI category. This suggests that individuals with lower daily steps within this range may have an increased likelihood of being classified as overweight.

The observation that daily steps below 4000 are associated with the "Obese" BMI category implies that individuals with lower physical activity levels may be more prone to obesity. This finding underscores the importance of regular physical activity in weight management.

These associations have practical implications for public health interventions and individual well-being. Encouraging higher daily step counts may be a beneficial strategy for weight management and reducing the risk of obesity.

```python
sns.displot(df, x='Heart Rate', y='Physical Activity Level', hue= 'Gender', kind='kde')
plt.title('Distance Plot for Representing Relation Between Physical Activity level and Heart Rates')
plt.show()
```

The observed overlap of clusters around heart rate 65 and physical activity level 70 between both genders implies that there is a common range of heart rate and physical activity level where individuals from both genders tend to cluster

This finding may suggest that individuals with heart rates around 70 and engaging in physical activity within the specified range may share common cardiovascular fitness levels. A heart rate of 70 is often considered within the normal range and may be reflective of efficient cardiovascular function.

The presence of outliers with heart rates between 80-90 for males and physical activity levels between 60-80 suggests that there are individuals within the male group who exhibit higher heart rates while engaging in moderate to high physical activity.

```python
plt.figure(figsize=(12, 6))
sns.histplot(df, x='Occupation', hue='Sleep Disorder')
plt.xticks(rotation=45)
plt.title('Histogram Representing Sleep Disorders For People with Different Occupations')
plt.show()
```

Most occupations show a low prevalence of sleep disorders, indicating that the majority of people surveyed do not report significant sleep-related issues.The histogram reveals that occupations such as Nurses, Teachers, and Salespeople stand out with a higher prevalence of sleep disorders.
This could be attributed to the nature of these professions, which often involve irregular working hours, high stress levels, and increased responsibility.

## Aggregation of Categorical Data

```python
sleep_disorder_count = df['Sleep Disorder'].value_counts()
colors = sns.color_palette("viridis", len(sleep_disorder_count))
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(sleep_disorder_count, labels=sleep_disorder_count.index, autopct='%1.1f%%', startangle=90, colors=colors)
ax.axis('equal')
plt.title('Distribution of Sleep Disorders')
plt.show()
```

The distribution highlights that while a substantial proportion of individuals report sleep disorders, a majority of the dataset does not exhibit significant sleep-related issues.

```python
plt.figure(figsize=(14, 8))
scatterplot=sns.scatterplot(x='Age', y='Heart Rate', hue='Stress Level', size='Sleep Duration', sizes=(20, 200), data=df, palette='deep', legend='full')
scatterplot.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Scatterplot with Multiple Visual Encodings')
plt.show()
```

A noticeable trend emerges among individuals above the age of 50. These individuals tend to exhibit lower stress levels and enjoy a good amount of sleep duration. This could be indicative of a more relaxed lifestyle or reduced work-related stress in the later stages of life.

In the age range of 25 to 45, a cluster of points with heart rates between 68 and 75 stands out. Within this cluster, there is a pattern of higher stress levels and medium sleep duration. This suggests that individuals in this age range, with a specific heart rate range, tend to experience elevated stress levels despite achieving only moderate sleep durations.

```python
sns.jointplot(df, x='Age', y="Sleep Duration", kind='kde', hue='Gender', fill=True)
plt.title("Contour Plot Showcasing Sleep Duration of Different Age Group Filtering Genders", y=1.25)
plt.show()
```

The data is filtered by gender, allowing for a comparison between male and female sleep duration patterns across different age groups.

A distinctive observation is the contour line for females above the age of 45. In this group, there is a minimum sleep duration of 6.5 hours. This suggests that a significant number of females in this age category tend to prioritize and achieve a minimum of 6.5 hours of sleep per night.This pattern might be indicative of health-conscious behavior or the recognition of the importance of sufficient sleep for well-being among females in this age group.

In contrast, the contour plot for males in the same age group does not exhibit a similar minimum sleep duration pattern. This could suggest that, on average, males in this age range may not prioritize a minimum of 6.5 hours of sleep as consistently as their female counterparts.

```python
sns.catplot(df, x ='Gender', y='Physical Activity Level', kind="violin")
plt.title("Comparison of Physical Activity Levels Of Both Genders Using Violin Plot")
plt.show()
```

The violin plot depicting physical activity levels for both genders reveals distinct patterns in the distribution of activity levels. While the overall shapes of the male and female violins are similar, the notable difference lies in the width of the violins. The narrower width of the female violin indicates a higher concentration of data points within a specific range of physical activity levels. This suggests that females in the dataset exhibit a more consistent pattern in their engagement with physical activity, with a majority falling within a narrower band.


The wider width of the male violin suggests a greater variability in physical activity levels among males. The broader distribution implies that males in the dataset demonstrate a more diverse range of activity levels, with some engaging in higher levels of physical activity while others exhibit lower levels.
